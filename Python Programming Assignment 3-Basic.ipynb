
# 1 Write a program to check the number of Positive or Negative

```python
A=int(input(" Enter the Value of A "))

if A>0:
    print("The Number {}  is a Positive NUmber".format(A))
elif A<0:
    print(" The Number {} is a Negative Number ".format(A))
else :
    print("The Number {} is equal to Zero".format(A))
```

# 2 Check whether the number is ODD or Even?

```python
def ops(n):
    for i in range(1,n):
        if i%2==0:
            print(" The number {} is Even".format(i))
        else:
            print( " The number {} is Odd".format(i))
ops(10)
```
